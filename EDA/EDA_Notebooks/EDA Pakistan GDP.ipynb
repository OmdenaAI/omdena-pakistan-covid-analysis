```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cpi = pd.read_csv('./New CPI_Data.csv')
```

```python
cpi.head()
```

```python
cpi.shape
```

```python
gdp = pd.read_csv('./pakistan gdp data.csv')
```

```python
gdp.head()
```

```python
gdp.shape
```

```python
unemployment = pd.read_csv('./Pakistan_unemployment_data.csv')
```

```python
unemployment.head()
```

```python
unemployment.shape
```

```python
df = cpi.merge(gdp, on = 'Year', how = 'outer', suffixes = ('_CPI', '_GDP'))
```

```python
merge = df.merge(unemployment, on = 'Year', how = 'outer')
```

```python
merge.shape
```

```python
merge
```

```python
merge.drop(['Country Name_GDP', 'Country Code_GDP', 'Country Name_CPI','Country Code_CPI'], 
          axis = 1, inplace = True)
```

```python
merge.head()
```

```python
col_names = ['Year', 'Country Name', 'Country Code', 'CPI Value', 'GDP value',
            'Unemployment rate', 'Indicator Name_CPI', 'Indicator Name_GDP',
            'Indicator Name']
```

```python
merge = merge.reindex(columns = col_names)
```

```python
merge.head()
```

```python
merge['Indicator Name'][0]
```

```python
merge.isnull().sum()
```

```python
fig, axes = plt.subplots(3, 1, sharex = False, figsize = (20,20))

fig.suptitle('Pakistan Economical Growth and Unemployment Analysis', fontsize = 20,
            verticalalignment = 'bottom')


#Consumer Price index
sns.set_style('darkgrid')
sns.barplot(ax = axes[0], x = merge['Year'], y = merge['CPI Value'])
axes[0].set_title('Inflation, consumer prices (annual %)')

# Gross Domestic Product

sns.barplot(ax = axes[1], x = merge['Year'], y = merge['GDP value'])
axes[1].set_title('GDP growth (annual %)')

# Unemployment

sns.barplot(ax = axes[2], x = merge['Year'], y = merge['Unemployment rate'])
axes[2].set_title('Unemployment, total (% of total labor force) (modeled ILO estimate)')

plt.show()
```

According to investopedia definition of GDP, Economists traditionally use gross domestic product (GDP) to measure economic progress. If GDP is rising, the economy is in solid shape, and the nation is moving forward. On the other hand, if gross domestic product is falling, the economy might be in trouble, and the nation is losing ground. 

As we see the data trend from 1991 to 2020, the GDP starts to from 2019 when Covid19 pandemic arises, which highly impact the economy of pakistan. Unemployment rate increasing with the passage of time, specially after the covid19 pandemic it increased.
